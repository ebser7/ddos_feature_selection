# Performance

### Helper Function

```python
def table_performance(pfm, method, decimals=2):
    means = (pfm.mean()*100).round(decimals)
    std = (pfm.std()*100).round(decimals)

    columns = ["Accuracy", "Precision", "Recall", "F1_Score"]

    means = means[columns]
    std = std[columns]

    df = pd.DataFrame()
    df["Mean"] = means.astype('string')
    df["Std"] = std.astype('string')
    df[method] = df["Mean"] + ' ± ' + df["Std"]
    df = pd.DataFrame(df[method])
    df = df.T

    return df
```

### Tables

```python
binary = pd.concat([table_performance(b_pfm_pre, "Preprocessing", 2),
                   table_performance(b_pfm_basic, "Basic Methods", 2),
                   table_performance(b_pfm_corr, "Correlation", 2)])
binary.T
```

```python
multiclass = pd.concat([table_performance(m_cat_pfm_pre, "Preprocessing", 1),
                   table_performance(m_cat_pfm_basic, "Basic Methods", 1),
                   table_performance(m_cat_pfm_corr, "Correlation", 1)])
multiclass.T
```

```python
ovo = pd.concat([table_performance(m_ovo_pfm_pre, "Preprocessing", 1),
                   table_performance(m_ovo_pfm_basic, "Basic Methods", 1),
                   table_performance(m_ovo_pfm_corr, "Correlation", 1)])
ovo.T
```

```python
ovr = pd.concat([table_performance(m_ovr_pfm_pre, "Preprocessing", 1),
                   table_performance(m_ovr_pfm_basic, "Basic Methods", 1),
                   table_performance(m_ovr_pfm_corr, "Correlation", 1)])
ovr.T
```

# Runtimes

### Helper Function

```python
def table_times(pfm, method, decimals=4):
    means = pfm.mean().round(decimals)
    std = pfm.std().round(decimals)

    columns = ["Fit_Time", "FS_Time", "Train_Time", "Predict_Time", "Score_Time"]

    means = means[columns]
    std = std[columns]

    df = pd.DataFrame()
    df["Mean"] = means.astype('string')
    df["Std"] = std.astype('string')
    df[method] = df["Mean"] + ' ± ' + df["Std"]
    df = pd.DataFrame(df[method])
    df = df.T

    return df
```

### Tables
