# Loading Data

### Multiclass Classifiers (One vs One)

```python
ovo_pre = pd.read_pickle(root + ovo_ind + "pre.pkl")
ovo_basic = pd.read_pickle(root + ovo_ind + "basic.pkl")
ovo_corr = pd.read_pickle(root + ovo_ind + "corr.pkl")
```

# Graphs

### Helper Function

```python
def load_metric(classifier, metric):
    if classifier == "Binary":
        metrics = pd.DataFrame([bin_pre[metric],
                            bin_basic[metric],
                            bin_corr[metric]])
    elif classifier == "Multiclass":
        metrics = pd.DataFrame([mul_pre[metric],
                            mul_basic[metric],
                            mul_corr[metric]])
    elif classifier == "OvR":
        metrics = pd.DataFrame([ovr_pre[metric],
                    ovr_basic[metric],
                    ovr_corr[metric]])
    else:
        metrics = pd.DataFrame([ovo_pre[metric],
                            ovo_basic[metric],
                            ovo_corr[metric]])

    metrics = metrics.T
    metrics.columns = ["Preprocessing", "Basic_Methods", "Correlation"]

    return metrics
```

## Accuracy

```python
bin_accuracies = load_metric("Binary", "Accuracy")
mul_accuracies = load_metric("Multiclass", "Accuracy")
ovr_accuracies = load_metric("OvR", "Accuracy")
ovo_accuracies = load_metric("OvO", "Accuracy")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Accuracy')

sns.boxplot(ax = axes[0, 0], data=bin_accuracies)
axes[0, 0].set_title('Binary Classifiers')

sns.boxplot(ax = axes[0, 1], data=mul_accuracies)
axes[0, 1].set_title('Multiclass Classifiers')

sns.boxplot(ax = axes[1, 0], data=ovr_accuracies)
axes[1, 0].set_title('Multiclass Classifiers (OvR)')

sns.boxplot(ax = axes[1, 1], data=ovo_accuracies)
axes[1, 1].set_title('Multiclass Classifiers (OvO)')

fig.tight_layout()
```

## Feature Selection vs Training Times

```python
bin_times = pd.DataFrame()
bin_times["Feature Selection"] = load_metric("Binary", "FS_Time").mean()
bin_times["Classifier Training"] = load_metric("Binary", "Train_Time").mean()
mul_times = pd.DataFrame()
mul_times["Feature Selection"] = load_metric("Multiclass", "FS_Time").mean()
mul_times["Classifier Training"] = load_metric("Multiclass", "Train_Time").mean()
mul_times = pd.DataFrame()
mul_times["Feature Selection"] = load_metric("Multiclass", "FS_Time").mean()
mul_times["Training"] = load_metric("Multiclass", "Train_Time").mean()
ovr_times = pd.DataFrame()
ovr_times["Feature Selection"] = load_metric("OvR", "FS_Time").mean()
ovr_times["Training"] = load_metric("OvR", "Train_Time").mean()
ovo_times = pd.DataFrame()
ovo_times["Feature Selection"] = load_metric("OvO", "FS_Time").mean()
ovo_times["Training"] = load_metric("OvO", "Train_Time").mean()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Feature Selection and Classifier Training Runtimes')

bin_times.plot.bar(ax = axes[0], stacked=True, rot=0)
axes[0].set_title('Binary Classifiers')
axes[0].set_ylabel('Runtime (seconds)')

mul_times.plot.bar(ax = axes[1], stacked=True, rot=0)
axes[1].set_title('Multiclass Classifiers')
axes[1].set_ylabel('Runtime (seconds)')

fig.tight_layout()
plt.savefig("../../pictures/vector/03_fit_times_share_independent.svg")
```
