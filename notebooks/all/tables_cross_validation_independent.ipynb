# Performance

### Helper Function

```python
def table_performance(pfm, method, decimals=2):
    means = (pfm.mean()*100).round(decimals)
    std = (pfm.std()*100).round(decimals)

    columns = ["Accuracy", "Precision", "Recall", "F1_Score"]

    means = means[columns]
    std = std[columns]

    df = pd.DataFrame()
    df["Mean"] = means.astype('string')
    df["Std"] = std.astype('string')
    df[method] = df["Mean"] + ' ± ' + df["Std"]
    df = pd.DataFrame(df[method])
    df = df.T

    return df
```

### Tables

```python
binary = pd.concat([table_performance(bin_pre, "Preprocessing", 2),
                   table_performance(bin_basic, "Basic Methods", 2),
                   table_performance(bin_corr, "Correlation", 2)])
binary.T
```

```python
multiclass = pd.concat([table_performance(mul_pre, "Preprocessing", 1),
                   table_performance(mul_basic, "Basic Methods", 1),
                   table_performance(mul_corr, "Correlation", 1)])
multiclass.T
```

```python
ovr = pd.concat([table_performance(ovr_pre, "Preprocessing", 1),
                   table_performance(ovr_basic, "Basic Methods", 1),
                   table_performance(ovr_corr, "Correlation", 1)])
ovr.T
```

```python
ovo = pd.concat([table_performance(ovo_pre, "Preprocessing", 1),
                   table_performance(ovo_basic, "Basic Methods", 1),
                   table_performance(ovo_corr, "Correlation", 1)])
ovo.T
```

# Runtimes

### Helper Function

```python
def table_times(pfm, method, decimals=4):
    means = pfm.mean().round(decimals)
    std = pfm.std().round(decimals)

    columns = ["Fit_Time", "FS_Time", "Train_Time", "Predict_Time", "Score_Time"]

    means = means[columns]
    std = std[columns]

    df = pd.DataFrame()
    df["Mean"] = means.astype('string')
    df["Std"] = std.astype('string')
    df[method] = df["Mean"] + ' ± ' + df["Std"]
    df = pd.DataFrame(df[method])
    df = df.T

    return df
```

### Tables

```python
binary = pd.concat([table_times(bin_pre, "Preprocessing", 4),
                   table_times(bin_basic, "Basic Methods", 5),
                   table_times(bin_corr, "Correlation", 3)])
binary
```

```python
multiclass = pd.concat([table_times(mul_pre, "Preprocessing", 4),
                   table_times(mul_basic, "Basic Methods", 4),
                   table_times(mul_corr, "Correlation", 4)])
multiclass
```

```python
ovr = pd.concat([table_times(ovr_pre, "Preprocessing", 4),
                   table_times(ovr_basic, "Basic Methods", 4),
                   table_times(ovr_corr, "Correlation", 4)])
ovr
```
